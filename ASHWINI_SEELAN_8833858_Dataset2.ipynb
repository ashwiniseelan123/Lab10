```python
# task-3.a

import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Print data types of all columns in the DataFrame
print(df_AGV.dtypes)

# Identify columns with non-numeric data types
non_numeric_columns = df_AGV.select_dtypes(exclude=['float64', 'int64']).columns
print("Non-numeric columns:", non_numeric_columns)

# Handling missing values: Fill missing values in numerical columns with the mean value
numerical_columns = ['B1.1  Net Balance brought forward from previous year',
                     'B2.1.1  Public Library Operating Grant (PLOG)',
                     'B2.1.2  Pay Equity Grant'
                    ]

# Convert numerical columns to numeric type
df_AGV[numerical_columns] = df_AGV[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Fill missing values with the mean of the respective column
for column in numerical_columns:
    mean_value = df_AGV[column].mean()
    df_AGV[column].fillna(mean_value, inplace=True)

# Drop non-numeric columns from numerical_columns
numerical_columns = [col for col in numerical_columns if df_AGV[col].dtype in ['float64', 'int64']]

# Normalize numerical features: Standardize numerical columns
if not df_AGV[numerical_columns].empty:
    scaler = StandardScaler()
    df_AGV[numerical_columns] = scaler.fit_transform(df_AGV[numerical_columns])

# Drop 'Unnamed: 309' column if it exists in the DataFrame
if 'Unnamed: 309' in df_AGV.columns:
    df_AGV.drop(['Unnamed: 309'], axis=1, inplace=True)

# Print the first few rows of the preprocessed DataFrame
print(df_AGV.head())

# Save the preprocessed DataFrame to a new CSV file (if needed)
df_AGV.to_csv("C:/Users/ashwi/Assignment10/preprocessed_2019_ontario_public_library_statistics_open_data.csv", index=False)
df_preprocessed_dataset_AGV = pd.read_csv("C:/Users/ashwi/Assignment10/preprocessed_2019_ontario_public_library_statistics_open_data.csv")
print(df_preprocessed_dataset_AGV.head())
```

```python
#####task4.A
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.multioutput import MultiOutputClassifier

# Assuming df_AGV contains the preprocessed DataFrame

# Define the feature columns (excluding the target column)
feature_columns = ['B1.1  Net Balance brought forward from previous year',
                   'B2.1.1  Public Library Operating Grant (PLOG)',
                   'B2.1.2  Pay Equity Grant'
                   ]

# Define the target column
target_column = ['F2.1.P  No. of programs held annually', 'F2.2.A  Annual program attendance']  # Replace 'Your_Target_Column_Name' with the actual column name

# Check for missing values in the feature columns
print(df_AGV[feature_columns].isnull().sum())

# Handle missing values in the feature columns using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # You can use 'median' or 'most_frequent' as well
df_AGV[feature_columns] = imputer.fit_transform(df_AGV[feature_columns])

# Check for missing values in the target columns
print(df_AGV[target_column].isnull().sum())

# Fill missing values with 0
df_AGV[target_column] = df_AGV[target_column].fillna(0)

# Convert non-numeric values to numeric format (float) and replace ',' with ''
df_AGV[target_column] = df_AGV[target_column].apply(lambda x: pd.to_numeric(x.astype(str).str.replace(',', ''), errors='coerce'))

# Handle missing values in the target columns using SimpleImputer
target_imputer = SimpleImputer(strategy='mean')  # You can use 'median' or 'most_frequent' as well
df_AGV[target_column] = target_imputer.fit_transform(df_AGV[target_column])

# Split the data into training and testing sets
X = df_AGV[feature_columns]
y = df_AGV[target_column]

# Re-split the data with the same random state to ensure consistency
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Fit the model to the data
rf_classifier.fit(X_train, y_train)

# Initialize a list to store accuracy scores for each target column
rf_accuracy = []

# Get the predictions for all target columns
y_pred_all_columns = rf_classifier.predict(X_test)

# Evaluate accuracy score for each target column separately
for i, col in enumerate(y_test.columns):
    y_true_col = y_test[col]
    y_pred_col = y_pred_all_columns[:, i]  # Get the predictions for the specific target column
    accuracy_col = accuracy_score(y_true_col, y_pred_col)
    rf_accuracy.append(accuracy_col)

print("Random Forest Model Accuracy for each target column:")
print(pd.Series(rf_accuracy, index=y_test.columns))


# 2. Support Vector Machine (SVM) Model using MultiOutputClassifier
svm_classifier = MultiOutputClassifier(SVC(kernel='linear', C=1.0, random_state=42))
svm_classifier.fit(X_train, y_train)

# Make predictions using the SVM model
svm_predictions = svm_classifier.predict(X_test)

# Initialize a list to store accuracy scores for each target column
svm_accuracy = []

# Evaluate accuracy score for each target column separately
for i, col in enumerate(y_test.columns):
    y_true_col = y_test[col]
    y_pred_col = svm_predictions[:, i]  # Get the predictions for the specific target column
    accuracy_col = accuracy_score(y_true_col, y_pred_col)
    svm_accuracy.append(accuracy_col)

print("SVM Model Accuracy for each target column:")
print(pd.Series(svm_accuracy, index=y_test.columns))
```
